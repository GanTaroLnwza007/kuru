## Explainability Approach

KUru is a RAG system, so the most interpretable model signals are retrieval evidence signals rather than neural-network weights. We explain model behavior at two levels:

| Level | Explanation signal | Why it is meaningful |
|-------|--------------------|----------------------|
| Retrieval | Top retrieved chunks and cosine similarity | Shows which official PDF passages drive the answer |
| Reranking | Keyword/section overlap for structure, PLO, policy, and TCAS queries | Shows why a chunk was promoted after dense search |
| Experiment | Change in good-answer rate across retrieval settings | Shows which hyperparameters most affect answer quality |

For non-technical users, this appears as source citations and confidence labels. For the engineering team, it appears as similarity bars, section-type breakdowns, and delta-from-baseline charts.

## Key Findings

| Finding | Interpretation |
|---------|----------------|
| More context can help, but is not enough by itself | Earlier `wider_retrieval` experiments improved some answers, but later v7 gains came from targeted lexical reranking rather than simply increasing `top_k`. |
| Dense table rows remain hard to explain and retrieve | Credit structures and assessment tables need table-aware chunking and section tags. |
| PLO and policy questions benefit from explicit lexical signals | Terms such as PLO, assessment, policy, plan numbers, and credit categories should influence reranking. |
| Source citations are the end-user explanation | Students do not need embeddings or similarity scores; they need the answer plus the official PDF source. |
